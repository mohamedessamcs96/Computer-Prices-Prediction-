```python
import matplotlib.pyplot as plt 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC 


%matplotlib inline
```

```python
computers=pd.read_csv('Computers.csv', delimiter=',')
computers.head()
```

```python
# Find missing values
print('Missing values:{}'.format(computers.isnull().any().sum()))
# Find duplicated records
print('\nNumber of duplicated records: {}'.format(computers.duplicated().sum()))

# Find the unique values of 'diagnosis'.
```

```python
# Review number of columns of each data type in a DataFrame:
computers.get_dtype_counts()
```

```python
computers.shape
```

```python
computers.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
computers.head()
```

```python
computers['cd'] = computers['cd'].map({'no':0,'yes':1})
computers.head()
```

```python
computers['premium'] = computers['premium'].map({'no':0,'yes':1})
computers.head()
```

```python
computers['multi'] = computers['multi'].map({'no':0,'yes':1})
computers.head()
```

```python
from sklearn.linear_model import LogisticRegression

X=computers[['speed','hd','ram','screen','cd','multi','premium','ads','trend']]

y=computers['price']

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)

log_reg=LogisticRegression(C=50)
log_reg.fit(X_train,y_train)
```

```python
predict=log_reg.predict(X_test)

predict
```
